```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ttest_ind
from scipy.stats import f_oneway
from scipy.stats import chi2_contingency
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
#1
heart = pd.read_csv("heart_disease.csv")
print(heart.head())
```

```python
#2
sns.boxplot(x=heart.thalach, y=heart.heart_disease)
```

```python
#3
thalac_hd = heart.thalach[heart["heart_disease"] == "presence"]
thalac_no_hd = heart.thalach[heart["heart_disease"] == "absence"]
```

```python
#4
print(np.mean(thalac_no_hd) - np.mean(thalac_hd))
print(np.median(thalac_no_hd) - np.median(thalac_hd))
```

```python
#5
statistic, pval = ttest_ind(thalac_no_hd, thalac_hd)
print(pval)

#6 pval < 0.05 -> significant difference
```

```python
#7
plt.clf()
sns.boxplot(x=heart.age, y=heart.heart_disease)
age_hd = heart.age[heart["heart_disease"] == "presence"]
age_no_hd = heart.age[heart["heart_disease"] == "absence"]

print(np.mean(age_hd) - np.mean(age_no_hd))
statistic, pval = ttest_ind(age_hd, age_no_hd)
print("pval is: ", pval)
```

```python
#7 Define formula and use it for chol
#quantitative & binary categorical
def var_irt_heartdisease(var):
    plt.clf()
    sns.boxplot(x=var, y=heart.heart_disease)
    with_hd = var[heart["heart_disease"] == "presence"]
    without_hd = var[heart["heart_disease"] == "absence"]

    print(np.mean(with_hd) - np.mean(without_hd))
    statistic, pval = ttest_ind(with_hd, without_hd)
    if pval < 0.05:
        print("The result has statistic significance")
        print("pval is: ", pval)
    
var_irt_heartdisease(heart.chol)
```

```python
print(heart["cp"].unique())
```

```python
#8 Formula for one quantitative + one categorical (non-binary)
#11 part B
def var_irt_var2(var, var2):
    plt.clf()
    for val in var2.unique():
        sns.boxplot(x=var, y = var2[var2==val])
        plt.show()
    
    tukey = pairwise_tukeyhsd(endog=var,
                          groups=var2,
                          alpha=0.05)
    print(tukey)

    
var_irt_var2(heart["thalach"], heart["cp"])
```

```python
thalach_typical = heart.thalach[heart.cp == 'typical angina']
thalach_asymptom = heart.thalach[heart.cp == 'asymptomatic']
thalach_nonangin = heart.thalach[heart.cp == 'non-anginal pain']
thalach_atypical = heart.thalach[heart.cp == 'atypical angina']
```

```python
#11
Fstat, pval = f_oneway(thalach_typical, thalach_asymptom, thalach_nonangin, thalach_atypical)
print('p-value for ANOVA: ', pval)
```

```python
#12
Xtab = pd.crosstab(heart["cp"], heart["heart_disease"])
print(Xtab)
```

```python
#13
chi2, pval, dof, expected = chi2_contingency(Xtab)
print(pval)
```
